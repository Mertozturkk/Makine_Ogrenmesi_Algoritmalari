```python
Y = y.reshape(-1, 1)
```

```python
def prediction(X, w):
    return np.matmul(X,w)
def compute_loss(X, Y, w):
    return np.average((prediction(X,w) - Y)**2)
def compute_gradient(X, Y, w):
    return 2*np.matmul(X.T, (prediction(X,w) - Y)) / X.shape[0]
def fit(X, Y, max_iter, learning_rate):
    weights = np.zeros((X.shape[1], 1))
    initial_loss = compute_loss(X, Y, weights)
    for i in range(max_iter):
        print("Iterasyon: {} Kayip: {:.5f}".format(i, compute_loss(X, Y, weights)))
        weights -= compute_gradient(X, Y, weights) * learning_rate  
        c_loss = compute_loss(X, Y, weights)
        if abs(initial_loss - c_loss) < 0.00001:
            return weights
        
        initial_loss = c_loss
    raise Exception("dÄ±dÄ±dÄ± dÄ±m")
```
